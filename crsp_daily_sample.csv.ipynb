```python
import pandas as pd

# 读取你电脑里的大文件
crsp = pd.read_csv('data/crsp_daily.csv')

# 只看有哪些股票
all_stocks = crsp['permno'].unique()
print(f"总共有 {len(all_stocks)} 只股票")

# 只取前 3 只股票的数据
small_stocks = all_stocks[:3]
crsp_small = crsp[crsp['permno'].isin(small_stocks)]

# 保存成小文件
crsp_small.to_csv('data/crsp_daily_sample.csv', index=False)

print(f"小文件已生成，形状：{crsp_small.shape}")
print(f"小文件大概 {crsp_small.memory_usage(deep=True).sum() / 1024 / 1024:.1f} MB")

```
